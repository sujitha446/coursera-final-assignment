## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
Q1=df['ConvertedComp'].quantile(0.25)
Q3=df['ConvertedComp'].quantile(0.75)
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
IQR=Q3-Q1
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers = df['ConvertedComp'][(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
if outliers.empty:
    print("No outliers found.")
else:
    print("Outliers found:")
    print(outliers)
ouliters_remove_list=outliers.index.values.tolist()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
new_df=df.drop(ouliters_remove_list).reset_index()
new_df
```

```python
median_value=new_df['ConvertedComp'].median()
median_value
```

```python
mean_value=new_df['ConvertedComp'].mean()
mean_value
```
