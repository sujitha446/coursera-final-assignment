```python
# Your code here
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

Download the webpage at the url


```python
#your code goes here

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html" # Replace with the URL you want to scrape
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    html_content = response.content
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
    html_content = None
print(html_content)
```

Create a soup object


```python
data = requests.get(url).text
print(data)
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
table = soup.select('table')[0] 
print(table)



 
```

Scrape the `Language name` and `annual average salary`.


Save the scrapped data into a file named *popular-languages.csv*


```python
# your code goes here
table_df = pd.read_html(str(table))[0] 
table_df.columns = table_df.iloc[0]
table_df.drop(table_df.index[0],inplace=True)
# table_df.set_index('No.',inplace=True)
table_df
```

```python

extracted_df = table_df[['Language', 'Average Annual Salary']] 
extracted_df.to_csv('example.csv', index=False)
```
